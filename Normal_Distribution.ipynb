## Quantile-Quantile (QQ) Plot
The percentiles of a standard normal distribution plotted agaist the actual percentiles of sample data. If actually a normal dist. there will be a close to linear relationship

```r
sample_data <- c(1,1,2,3,3,4,5,4,4,5,5,5,6,6,7,7,8,7,7,8,9,10,11)
```

```r
# Normal Dist
qqnorm(sample_data)
```

```r
# Histogram
hist(sample_data)
```

Now viewing both sample_data visualizations we can determine this data is relatively normally distributed. 

### Compare QQ plot for first dataset to view distribution

```r
qqnorm(my_data)
```

This QQ plot coupled with the intial histogram reveals a non-linear relationship - flag this is not normal distribution

# Skewness and Kurtosis
##### How to quantify the data
##### Skewness: 
How asymmetric the distribution is compared to normal 
##### Kurtosis:
How pointed

```r
# Install moments library for related functions 
library(moments)
```

```r
skewness(my_data)
```

```r
skewness(sample_data)
```

We intrepret this as 'my_data' is moderately skewed (positive direction), this is supported in the histogram. Then for 'sample_data' it has a low skewness(0.115) - which would mean it is relitively symmetrically distributed which is consistent with the histogram and QQ plot

```r
kurtosis(my_data)
```

```r
kurtosis(sample_data)
```

##### Standard normal has kurtosis = 3
Less informative, both datasets are slightly flatter than a standard normal dist.

## Formal Tests for Normality

* Anderson-Darling
* Shapiro-Wilk
* Lilliefors

In all tests, the null and alternative hypothesis is same
H0 = random observations sampled from normally dist population
h1 = observations from non-normally dist

### Shapiro-Wilk
Most common test. 

```r
shapiro.test(my_data)
```

```r
shapiro.test(sample_data)
```

##### Conclusions 
'my_data' reject null hypothesis (significance lvl of .05) since p-lvl is .013 
'sample_data' retains null hypothesis when p-lvl = 0.8468, the data is normally distributed 

### Lilliefors 
Specifically for normal dist. 

```r
library(nortest)
```

```r
lillie.test(my_data)
```

```r
lillie.test(sample_data)
```

##### Lillefors is known to be more conservative but less precise for normality. In this example both datasets we retain the null hypothesis

```r
### Anderson-Darling 
ad.test(my_data)
```

```r
ad.test(sample_data)
```

##### Conclusion
Results similar to Shapiro-wilk we reject null for my_data and retain for sample_data
